```python
## We need to do label encoding to convert object into integer
## use def to define new function and use return statement to make function send objects back to caller code

from sklearn.preprocessing import LabelEncoder

def label_encoded(encode):
     le = LabelEncoder()
     le.fit(encode)
     return le.transform(encode)
for col in df.columns:
    df[str(col)] = label_encoded(df[str(col)])
```

```python
##  Remove less corrleted data
df.drop(columns=['veil-type', 'gill-attachment', 'veil-color', 'stalk-root'], axis=1, inplace = True)
```

```python
## Select variables (X,Y)
X = df.iloc[:,1:]
Y = df['class']
```

```python
## split the data into training ( 80%) and testing set (20%)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size= 0.2, random_state = 32)
```

```python
## standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## Build Decision tree model
from sklearn.tree import DecisionTreeClassifier
# Create default Decision Tree classifer object
decisionTree_model_gini = DecisionTreeClassifier()
decisionTree_model_gini.fit(X_train,Y_train)
```

```python
# Use entropy criterion
decisionTree_model_entropy = DecisionTreeClassifier(criterion="entropy")
decisionTree_model_entropy.fit(X_train,Y_train)
```

```python
# Use max depth of 10
decisionTree_model_depth = DecisionTreeClassifier(max_depth=10)
decisionTree_model_depth.fit(X_train,Y_train)
```

```python
## Make Prediction
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
from mlxtend.plotting import plot_confusion_matrix

Y_pred_gini = decisionTree_model_gini.predict(X_test)
Y_pred_entropy = decisionTree_model_entropy.predict(X_test)
Y_pred_depth = decisionTree_model_depth.predict(X_test)
```

```python
## Classification Model Evaluataion  (gini)
accuracy_gini = accuracy_score(Y_test, Y_pred_gini)
conf_mat_gini = confusion_matrix(Y_test, Y_pred_gini)
print(accuracy_gini)
print(conf_mat_gini)
## Classification Model Evaluataion  (entropy)
accuracy_entropy = accuracy_score(Y_test, Y_pred_entropy)
conf_mat_entropy = confusion_matrix(Y_test, Y_pred_entropy)
print(accuracy_entropy)
print(conf_mat_entropy)
## Classification Model Evaluataion (max_depth)
accuracy_max_depth = accuracy_score(Y_test, Y_pred_depth)
conf_mat_max_depth = confusion_matrix(Y_test, Y_pred_depth)
print(accuracy_max_depth)
print(conf_mat_max_depth)
```

```python
## plot confusion matrix
cm = confusion_matrix(Y_test, Y_pred_gini)
fig, ax = plot_confusion_matrix(conf_mat=cm, figsize=(4, 4))
plt.xlabel('Predictions', fontsize=10)
plt.ylabel('Actuals', fontsize = 10)
plt.title('Confusion Matrix', fontsize = 10)
plt.show()
```

```python
cm = confusion_matrix(Y_test, Y_pred_entropy)
fig, ax = plot_confusion_matrix(conf_mat=cm, figsize=(4, 4))
plt.xlabel('Predictions', fontsize=10)
plt.ylabel('Actuals', fontsize = 10)
plt.title('Confusion Matrix', fontsize = 10)
plt.show()
```

```python
print(classification_report(Y_test, Y_pred_gini))
```

```python
FPR, TPR = roc_curve(Y_test, Y_pred_gini)[0:2]
AUC_score = roc_auc_score(Y_test, Y_pred_gini)
plt.plot(FPR, TPR, color='orange', label="AUC="+str(AUC_score))
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Decision Tree-gini')
plt.legend()
plt.show()
```

```python
## K-Nearest Neighbours (KNN) Model with for loop [1-20] to check the best accuracy of n
from sklearn.neighbors import KNeighborsClassifier
for n in range(1, 21):
    knn_model = KNeighborsClassifier(n_neighbors = n)
    knn_model.fit(X_train, Y_train)
    Y_pred_knn = knn_model.predict(X_test)
    print('KNeighborsClassifier: n = {} , Accuracy is: {}'.format(n,knn_model.score(X_test,Y_test)))
```

```python
## Make prediction for KNN
knn_model = KNeighborsClassifier(n_neighbors = 10)
knn_model.fit(X_train, Y_train)
Y_pred_knn = knn_model.predict(X_test)
pd.DataFrame({'Actual': Y_test, 'Predicted':Y_pred_knn})
```

```python
## Check Confusion matrix
conf_mat_knn = confusion_matrix(Y_test, Y_pred_knn)
print(conf_mat_knn)
```
